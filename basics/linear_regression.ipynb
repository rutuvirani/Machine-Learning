```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/rutuvirani/Machine-Learning/refs/heads/main/basics/Data_LR.csv"
data = pd.read_csv(url) # Import data from csv file uploded on basics folder
data.head() 

x = data.iloc[:,0:1] # Separate column this is 1st column
y = data.iloc[:,-1]  # Indexing with -1 will choose last column

from sklearn.model_selection import train_test_split # we want to split our data in to train set and test set

# data divide into test size 0.2 means 20% test data and 80% train data
# random_state=2 means when we use this code as integer 2 then same randomization follows every time
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2) 

from sklearn.linear_model import LinearRegression

lr = LinearRegression() # made object
lr.fit(x_train,y_train) # This attribute .fit() will train model means it fine the best fitted line 

x_test.iloc[0]

# we can predict the y(package) from inputing x(cgpa), reshape() uses to change the shape of array
lr.predict(x_test.iloc[0].values.reshape(1,1)) 

plt.scatter(data['cgpa'],data['package(Mn)'])
plt.plot(x_train,lr.predict(x_train),color='red') # draw line with plt.plot((x1,y1),(x2,y2))
plt.xlabel('cgpa')
plt.ylabel('package(Mn)')

m = lr.coef_    # This attribute .coef_ will give us slope of line

b = lr.intercept_   #   This attribute .intercept_ will give us intercept of line 
```
