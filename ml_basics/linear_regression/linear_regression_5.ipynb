```python
# This is gradient linear regression with manual calculation of m and b, check ReadMe.md

# Data generation
import numpy as np
import matplotlib.pyplot as plt

m = 2.5          # Slope
b = 1.0          # Intercept
n = 10         # Number of data points
noise_level = 20  # Noise level (adjustable)

# Generate x-values (independent variable)
X = np.random.rand(n) * 100  # Random x values between 0 and 100

# Calculate y-values based on the linear relationship with added noise we add noice to make more real data
y = m * X + b + np.random.randn(n) * noise_level  # y = mx + b + noise, w/o noise it will be line

plt.scatter(X,y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
class GDRegressor:

    def __init__(self,learning_rate,epochs):
        self.m = 0 # initialize random m value
        self.b = 0 # initialize random b value
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,X,y): # To train model we need formula to find m and b
        for i in range(self.epochs):
            loss_slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b) # partial derivative of loss with respect to m and b
            loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.b) * X.ravel()) # or gradient of loss with respect to m and b
            self.m = self.m - (self.lr * loss_slope_m) # update m 
            self.b = self.b - (self.lr * loss_slope_b) # update b

            y_pred = self.m*X + self.b
            plt.plot(X,y_pred)
        plt.scatter(X,y)
    
        print(self.m,self.b)
        
    def predict(self,X):
        return self.m*X + self.b
    
```

```python
GDR = GDRegressor(0.00001, 1000)
GDR.fit(X,y)
# plt.scatter(X,y)
```
